```python
# ECUACIÓN DE ESTADO DE PENG ROBINSON PARA UNA MEZCLA BINARIA
import CoolProp.CoolProp as CP
import numpy as np
from scipy.optimize import newton
import requests
from PIL import Image
from io import BytesIO
```

```python
# Constante de los gases
R = 8.314472  # J /mol·K

print('-----ECUACIÓN DE ESTADO DE REDLICH-KWONG-----')

# Ingreso de los nombres de las sustancias
nombre_sustancia1 = input('Ingrese el nombre de la primera sustancia en inglés: ')
nombre_sustancia2 = input('Ingrese el nombre de la segunda sustancia en inglés: ')
print("")

# Función para obtener o ingresar manualmente los valores críticos
def obtener_tc_pc(nombre):
    try:
        TC = CP.PropsSI('Tcrit', nombre)
        PC = CP.PropsSI('Pcrit', nombre) * 1e-5  # Pa a bar
        print(f'Sustancia seleccionada: {nombre}')
        print(f'Temperatura crítica (TC): {TC} K')
        print(f'Presión crítica (PC): {PC} bar\n')
    except:
        print(f'No se encontró "{nombre}" en la base de CoolProp.')
        TC = float(input(f'Ingrese manualmente la temperatura crítica TC [K] para {nombre}: '))
        PC = float(input(f'Ingrese manualmente la presión crítica PC [bar] para {nombre}: '))
    return TC, PC

# Obtener datos críticos para ambas sustancias
TC1, PC1 = obtener_tc_pc(nombre_sustancia1)
TC2, PC2 = obtener_tc_pc(nombre_sustancia2)

# Fracciones molares
x1 = float(input('\nIngrese la fracción molar de la primera sustancia: '))
x2 = 1 - x1

# Parámetros de Redlich-Kwong para cada sustancia
a1 = (9*((2**(1/3))-1))**(-1) * ((R**2 * TC1**2.5)/(PC1*1e5))
a1 = round(a1, 5)
b1 = (((2**(1/3))-1)/3) * ((R * TC1) / (PC1*1e5))

a2 = (9*((2**(1/3))-1))**(-1) * ((R**2 * TC2**2.5)/(PC2*1e5))
a2 = round(a2, 5)
b2 = (((2**(1/3))-1)/3) * ((R * TC2) / (PC2*1e5))

# Parámetros de la mezcla utilizando reglas de mezclado
print("\n--- PARÁMETROS DE MEZCLA ---")
a = x1**2 * a1 + x2**2 * a2 + 2 * x1 * x2 * (a1 * a2)**0.5
b = x1 * b1 + x2 * b2

print(f' amix = {round(a, 5)} [(J·m³·K^0.5)/mol²]')
print(f' bmix = {format(b, "0.5e")} [m³/mol]')
```

```python
# Cálculo de volumen
P = float(input('Presión (en bar): '))
P = P * 100000
T = float(input('Temperatura (en Kelvin): '))

c1 = 1
c2 = -(R*T)/P
c3 = (a / (P * T**0.5)) - (b**2) - ((b * R * T) / P)
c4 = -(a*b)/(P * T**0.5)
coeff = [c1, c2, c3, c4]
v = np.roots(coeff)
print('v1 = ', np.round(v[0], 6), '[m^3/mol] ')
print('v2 = ', np.round(v[1], 8), '[m^3/mol] ')
print('v3 = ', np.round(v[2], 8), '[m^3/mol] ')
```
