```python
import CoolProp.CoolProp as CP
import numpy as np
import requests
from scipy.optimize import newton
import requests
from PIL import Image
from io import BytesIO
```

```python
# Constante de los gases
R = 8.314472  # J /mol·K

print('-----ECUACIÓN DE ESTADO DE REDLICH-KWONG-----')

# Ingreso de nombres de sustancias
nombre_sustancia1 = input('Ingrese el nombre de la primera sustancia en inglés: ')
nombre_sustancia2 = input('Ingrese el nombre de la segunda sustancia en inglés: ')
print("")

# Función para obtener o ingresar manualmente TC y PC
def obtener_parametros(sustancia):
    try:
        TC = CP.PropsSI('Tcrit', sustancia)
        PC = CP.PropsSI('Pcrit', sustancia) * 1e-5  # Pa a bar
        print(f'Datos obtenidos de CoolProp para {sustancia}')
    except Exception:
        print(f'No se encontraron datos críticos para "{sustancia}".')
        TC = float(input(f'Ingrese la temperatura crítica TC [K] para {sustancia}: '))
        PC = float(input(f'Ingrese la presión crítica PC [bar] para {sustancia}: '))
    
    # Cálculo de parámetros a y b
    a = (9*((2**(1/3))-1))**(-1) * ((R**2 * TC**2.5)/(PC*1e5))
    b = (((2**(1/3))-1)/3) * ((R * TC) / (PC*1e5))
    return a, b, TC, PC

# Obtener parámetros para ambas sustancias
a1, b1, TC1, PC1 = obtener_parametros(nombre_sustancia1)
print("")
a2, b2, TC2, PC2 = obtener_parametros(nombre_sustancia2)

# Mostrar resultados individuales
print("\n--- PARÁMETROS INDIVIDUALES ---")
print(f'{nombre_sustancia1}: a1 = {round(a1,5)} [(J·m³·K^0.5)/mol²], b1 = {format(b1,".5e")} [m³/mol]')
print(f'{nombre_sustancia2}: a2 = {round(a2,5)} [(J·m³·K^0.5)/mol²], b2 = {format(b2,".5e")} [m³/mol]')

# Fracciones molares
x1 = float(input('\nIngrese la fracción molar de la primera sustancia: '))
x2 = 1 - x1

# Mezcla
a = x1**2 * a1 + x2**2 * a2 + 2 * x1 * x2 * (a1 * a2)**0.5
b = x1 * b1 + x2 * b2
print("\n--- PARÁMETROS DE LA MEZCLA ---")
print('amix =', round(a, 5), '[(J·m³·K^0.5)/mol²]')
print('bmix =', format(b, '0.5e'), '[m³/mol]')
```

```python
# Cálculo de volumen
P = float(input('Presión (en bar): '))
P = P * 100000
T = float(input('Temperatura (en Kelvin): '))

c1 = 1
c2 = -(R*T)/P
c3 = (a / (P * T**0.5)) - (b**2) - ((b * R * T) / P)
c4 = -(a*b)/(P * T**0.5)
coeff = [c1, c2, c3, c4]
v = np.roots(coeff)
print('v1 = ', np.round(v[0], 6), '[m^3/mol] ')
print('v2 = ', np.round(v[1], 8), '[m^3/mol] ')
print('v3 = ', np.round(v[2], 8), '[m^3/mol] ')
```
