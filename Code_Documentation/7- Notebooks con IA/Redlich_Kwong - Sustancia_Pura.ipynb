```python
import CoolProp.CoolProp as CP
import numpy as np
import requests
from scipy.optimize import newton
import requests
from PIL import Image
from io import BytesIO
# Constante de los gases
R = 8.314472  # J /mol K

print('-----ECUACIÓN DE ESTADO DE REDLICH-KWONG-----')

# Selecciona la sustancia en la base de CoolProb (en ingles)
nombre_sustancia = input('Ingrese el nombre de la sustancia en ingles: ')

# Obtener propiedades críticas de la sustancia usando CoolProp
try:
    TC = CP.PropsSI('Tcrit', nombre_sustancia)
    PC = CP.PropsSI('Pcrit', nombre_sustancia) * 1e-5  # Convertir Pa a bar
    print(f'Sustancia seleccionada: {nombre_sustancia}')
    print(f'Temperatura crítica (TC): {TC} K')
    print(f'Presión crítica (PC): {PC} bar')
except Exception as e:
    print(f'Error: {e}')
    print('Sustancia no encontrada en la base de datos de CoolProp.')
    TC= float(input('Ingrese la temperatura critica de la sustancia en Kelvin : '))
    PC= float(input('Ingrese la Presion critica de la sustancia en bar : '))
    exit()

# Parámetros de la ecuación de Redlich-Kwong
a = (9*((2**(1/3))-1))**(-1) * ((R**2 * TC**2.5)/(PC*1e5))
a = round(a, 5)
b = (((2**(1/3))-1)/3) * ((R * TC) / (PC*1e5))
print(' a = ', a, ' [(J m^3 K^0.5) / mol^2] ')
print(' b = ', format(b, '0.5e'), ' [m^3 / mol] \n ')
# Cálculo de volumen
P = float(input('Presión (en bar): '))
P = P * 100000
T = float(input('Temperatura (en Kelvin): '))

c1 = 1
c2 = -(R*T)/P
c3 = (a / (P * T**0.5)) - (b**2) - ((b * R * T) / P)
c4 = -(a*b)/(P * T**0.5)
coeff = [c1, c2, c3, c4]
v = np.roots(coeff)
print('v1 = ', np.round(v[0], 6), '[m^3/mol] ')
print('v2 = ', np.round(v[1], 8), '[m^3/mol] ')
print('v3 = ', np.round(v[2], 8), '[m^3/mol] ')

from IPython.display import display, Markdown
from dotenv import load_dotenv
import os
load_dotenv("xxx.env")

api_key = os.getenv("OPENAI_API_KEY")

if api_key:
    print("Clave api encontrada, los primeros digitos son:", api_key[:8]+"..." + api_key[-4:])
else:
    print("No se encontro la clave API. Revisa tu archivo .env o la ruta.")

import openai
client=openai.OpenAI(api_key=api_key)
#Pedir al usuario una pregunta
pregunta=input("¿qué quieres preguntarle a la IA?:")
#Dar contexto a la IA
mensaje_encuenta= (
    f"Las raíces (volúmenes molares en m^3/mol) que se obtuvieron son: {v.tolist()}. "
    f"{pregunta}"
)
#Enviar pregunta a la IA
response = client.chat.completions.create(
model="gpt-4o",
messages= [
{"role":"system","content":"Eres un asistente útil, sabes de fisicoquimica y termodinamica"},
{"role":"user","content": mensaje_encuenta},
]
)
#Mostrar la respuesta
print(response.choices[0].message.content)
```
